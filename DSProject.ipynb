```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

###   Uses:-
     torch -> To load the YOLO model and make Detection
     matplotlib -> visualizing images
     cv2 -> to access web cam and render feeds
     numpy-> numerical calcullation


```python
# Model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)#load pretrained ultralytics model from torch hub
```

```python
# Images
imgs = ['https://i0.wp.com/calmatters.org/wp-content/uploads/2019/11/07222019_traffic_AW_sized_02.jpg?fit=1200%2C762&ssl=1'] 
```

# Detecting from Images

```python
# Inference
results = model(imgs)
# Results
results.print()
results.show()
```

# Detecting From video

```python
cap = cv2.VideoCapture('video.mp4')
while cap.isOpened():
    ret, frame = cap.read()
    # detection
    results = model(frame) 
    cv2.imshow('Object Detection', np.squeeze(results.render()))
    # to close the window
    if cv2.waitKey(10) & 0xFF == ord('q'):  
        break
cap.release()
cv2.destroyAllWindows()
        
```

# Detecting From web cam

```python
# access web camera
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    # detection
    results = model(frame) 
    cv2.imshow('Object Detection', np.squeeze(results.render()))
    # to close the window
    if cv2.waitKey(10) & 0xFF == ord('q'):  
        break
cap.release()
cv2.destroyAllWindows()
        
```
